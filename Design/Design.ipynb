# Design

## Defining the business problem to be solved

Many different products such as games, cooking, clothes, tech and more are released to the world for consumers to get their hands on, and their feedback is used to improve on their next iteration of the product. The issue is that some websites (such as Twitter/X, Facebook or really any social media), such as that of influencers, are unable to provide a quantifiable feedback. The service that I would develop is a website that someone could provide a transcript or just a written review, and see if the sentiment of the feedback is positive or negative (as a quantifiable value).

# Refactoring the business problem into a machine learning problem

The aim of this project is to use natural language processing in tandem with machine learning to teach a model what is considered positive and negative, and classify a sentiment based on it. By doing so, businesses are able to determine their product feedback in a form of either positive or negative. 

# Defining success metrics

Success is determined when the model is ableto accurately determine the sentiment, primarily through tests and real world applications/experiences. It is considered to fail if it does not classify at all with any sort of accuracy (such as 10% prediction rate being unacceptable, but 80% being not bad [not great either]). 

# Researching available data.

## Raw Data Sources

| Dataset | Source | Rows | Columns | Description |
| ------ | ------ | ------ | ------ | ------ |
| Dataset 1 | Kaggle Twitter Entity Sentiment Analysis | ~73,000 | 2 | Twitter posts with sentiment labels (Positive, Negative, Neutral, Irrelevant). Contains tweets about various brands and products, sourced from Twitter's public API. |
| Dataset 2 | Hugging Face Sentiment Analysis Complex | 500,000 (sampled) | 3 | Pre-curated sentiment analysis dataset from NNEngine with text content and sentiment labels. Provides a large, diverse collection of sentiment-labelled text data to improve model generalisation. |

## Column Descriptions (Raw Data)

| Column | Dataset | Data Type | Description |
| ------ | ------ | ------ | ------ |
| Text Content | Both | String | The raw text of the tweet or review. Contains user-generated feedback about products/services that may include typos, slang, emojis, and URLs. |
| Sentiment | Both | Categorical | Classification label: **Positive**, **Negative**, **Neutral** (Dataset 1); **positive**, **negative** (Dataset 2 after standardisation). Represents the overall sentiment expressed in the text. |
| id | Dataset 2 | Integer | Unique identifier for each record. Not used in model training but useful for tracking data provenance. |

## Data After Preprocessing

| Component | Details |
| ------ | ------ |
| Text Cleaning | Removed empty rows, duplicates, unknown tokens `<unk>`, and extra whitespace. Standardised sentiment labels to lowercase for consistency. |
| Tokenisation | Used NLTK TweetTokenizer to split text into individual tokens whilst preserving context (handles URLs, mentions, contractions). |
| Stopword Removal | Removed common words (a, the, and, or, etc.) but **preserved negations** (not, no, never) to maintain sentiment context. Example: "not good" stays intact rather than becoming "good". |
| Final Columns | **Text Content** (cleaned and tokenised), **Sentiment** (binary: positive/negative), **Mutated Text Content** (preprocessed version for feature extraction) |

## Feature Engineering Output

| Feature Type | Dimensions | Creation Method | Purpose |
| ------ | ------ | ------ | ------ |
| TF-IDF Features | 5,000 | TfidfVectorizer with bigrams (1-2 grams) | Captures word importance and frequency patterns. Weights common words lower and rare words higher. Enables model to identify which words correlate with positive vs. negative sentiment. |
| SBERT Embeddings | 128 | Sentence-BERT (all-MiniLM-L6-v2) | Captures semantic meaning and context of the entire text. Addresses TF-IDF's limitation of treating words in isolation. Combined with TF-IDF for hybrid approach. |
| Combined Features | 5,128 | Horizontal stack of TF-IDF + SBERT | Used as input to Logistic Regression model. Provides both interpretable token-level features and deep contextual understanding. |
